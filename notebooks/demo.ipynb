```python
import os
import torch
from scripts.train import train_model
import matplotlib.pyplot as plt
import pandas as pd

# Set up parameters for the study
data_dir = "/path/to/data"
data_file = "/path/to/data_file.txt"
task_name = "Task02_Heart"
epochs = 100
batch_size = 4
image_size = (64, 64, 32)
patch_size = (8, 8, 4)
lr = 3e-4
dataset_fraction = "full"
model_label = "ViT_Baseline"
```

```python
# Train the model
train_model(
    data_dir=data_dir,
    data_file=data_file,
    save_path=task_name,
    task_name=task_name,
    epochs=epochs,
    lr=lr,
    batch_size=batch_size,
    image_size=image_size,
    patch_size=patch_size,
    dataset_fraction=dataset_fraction,
    model_label=model_label
)
```

```python
# Load and visualize training results
results_path = os.path.join(task_name, f"{task_name}_training_results.csv")
df = pd.read_csv(results_path)

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(df['Epoch'], df['Train Loss'], label='Train Loss')
plt.plot(df['Epoch'], df['Val Loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Plot training and validation Dice score
plt.figure(figsize=(10, 5))
plt.plot(df['Epoch'], df['Train Dice'], label='Train Dice')
plt.plot(df['Epoch'], df['Val Dice'], label='Val Dice')
plt.xlabel('Epoch')
plt.ylabel('Dice Score')
plt.title('Training and Validation Dice Score')
plt.legend()
plt.grid(True)
plt.show()

# Plot training and validation NSD score
plt.figure(figsize=(10, 5))
plt.plot(df['Epoch'], df['Train NSD'], label='Train NSD')
plt.plot(df['Epoch'], df['Val NSD'], label='Val NSD')
plt.xlabel('Epoch')
plt.ylabel('NSD Score')
plt.title('Training and Validation NSD Score')
plt.legend()
plt.grid(True)
plt.show()
```
